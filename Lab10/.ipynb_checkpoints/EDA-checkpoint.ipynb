```python
df.drop_duplicates(inplace=True)  # Remove duplicate rows
#df = df.drop_duplicates()  # Alternative way to remove duplicates
```

```python
print(df.isnull().sum())  # Check for null values in each column
```

```python
#df.dropna(how='any', inplace=True)  # Remove rows with any null values
df.dropna(how='all', inplace=True)  # Remove rows with all null values
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna(df[col].mode()[0])  # Fill null values with mode for categorical columns
    else:
        df[col] = df[col].fillna(df[col].median())
```

```python
#df['Heart Disease'] = LabelEncoder().fit_transform(df['Heart Disease'])  # Encode target variable
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = LabelEncoder().fit_transform(df[col])  # Encode categorical columns
```

```python
columns_to_scale = ['Age','BP','Cholesterol','Max HR','ST depression']
#scaler = MinMaxScaler()
scaler = StandardScaler()
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])  # Scale specified columns
```

```python
correaltion_matriz = df.corr()  # Calculate correlation matrix

plt.figure(figsize=(30, 15))
sns.heatmap(correaltion_matriz, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.xticks(rotation=30,ha='right')
plt.show()
```

```python
df.drop(columns=['Age'],inplace=True)  # Drop 'Age' column
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Cholesterol'])
plt.title('Before Outlier Removal - Cholesterol')
plt.show()
```

```python
#IQR method for outlier removal
Q1 = df['Cholesterol'].quantile(0.25)
Q3 = df['Cholesterol'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['Cholesterol'] >= lower_bound) & (df['Cholesterol'] <= upper_bound)]  # Remove outliers
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Cholesterol'])
plt.title('After Outlier Removal - Cholesterol')
plt.show()
```

```python
print(df['Heart Disease'].value_counts())  # Check class distribution
```

```python
x = df.drop(columns=['Heart Disease'])  # Features
y = df['Heart Disease']  # Target variable
```

```python
print(f"Original Size: {y.value_counts()}")  # Print original class distribution
#balance = SMOTE(random_state=42)  # Initialize SMOTE
#balance = TomekLinks()  # Initialize Tomek Links
balance = SMOTETomek(random_state=42)  # Initialize SMOTETomek
x,y = balance.fit_resample(x, y)  # Resample the dataset
print(f"Resampled Size: {y.value_counts()}")  # Print resampled class distribution

```

```python
df = pd.concat([x, y], axis=1)  # Combine features and target variable back into a DataFrame
df.to_csv('Heart_Disease_Prediction_Processed.csv', index=False)  # Save the processed DataFrame to a new CSV file
```
